# Preprocessing

```python
df = data.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

```python
trainset['AveragePrice'].mean()
```

```python
testset['AveragePrice'].mean()
```

## Encoding, Feature engineering, imputing

```python
from sklearn.preprocessing import StandardScaler
```

```python
lab_enc_type = LabelEncoder()
lab_enc_year = LabelEncoder()
lab_enc_region = LabelEncoder()
```

```python
import warnings
warnings.simplefilter("ignore")
    
def encoding(df):
    # Todo : Check why I have this warning
    # https://www.dataquest.io/blog/settingwithcopywarning/
    df.loc[:, 'type'] = lab_enc_type.fit_transform(df['type'])
    df.loc[:, 'year'] = lab_enc_year.fit_transform(df['year'])
    df.loc[:, 'region'] = lab_enc_region.fit_transform(df['region'])
    return df
```

```python
def feature_engineering(df, columns):
    df = df.drop(['Unnamed: 0', 'Date'], axis=1)
    return df.drop(columns, axis=1)
```

```python
def imputation(df):
    # No need for this dataset, no 'nan' data
    return df
```

```python
def preprocessing(df, hypothesis):
    df = encoding(df)
    df = feature_engineering(df, hypothesis)
    df = imputation(df)
    
    X = df.drop('AveragePrice', axis=1)
    y = df['AveragePrice']

    return X,y
```

## Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold, cross_val_score
```

```python
n_folds = 5

def evaluation(model, hypothesis=[]):
    X_train, y_train = preprocessing(trainset, hypothesis)
    X_test, y_test = preprocessing(testset, hypothesis)
    
    cv_rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv = 5))
    print('Score : mean {:,.4f} std {:,.4f}'.format(cv_rmse.mean(), cv_rmse.std()))
```

## Test hypothesis on estimator

```python
from sklearn.model_selection import learning_curve
from sklearn.svm import SVR
```

```python
estimator = SVR(C=1.0, epsilon=0.2)
```

```python
hypothesis = [
    [],
    ['4046', '4225', '4770'],
    ['Small Bags', 'Large Bags', 'XLarge Bags'],
    ['4046', '4225', '4770', 'Small Bags', 'Large Bags', 'XLarge Bags'],
    ['4046', '4225', '4770', 'Small Bags', 'Large Bags', 'XLarge Bags', 'Total Volume', 'Total Bags']
]
```

```python
for i, hypo in enumerate(hypothesis):
    print('-----' + str(i) + '-----')
    evaluation(estimator, hypothesis=hypo)
```

The best score was the hypothesis 5 without phu and bags so I continue with it

## Test preprocessors

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures, RobustScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
preprocessors = {
    'PCA' : PCA(n_components=3),
    'SelectKBest 3': SelectKBest(f_classif, k=3),
    'RobustScaller': RobustScaler(),
    'PF + SKB': make_pipeline(PolynomialFeatures(3), SelectKBest(f_classif, k=3)),
    'StandardScale': StandardScaler()
}
```

```python
for name, preprocessor in preprocessors.items():
    print(name)
    evaluation(make_pipeline(preprocessor, estimator), hypothesis=hypothesis[-1])
```

RobustScaler seems to work fine, our dataset must have some outliars

## Estimator decision

```python
from sklearn.linear_model import Ridge, ElasticNet, SGDRegressor
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
```

```python
estimators = {
    'Ridge': Ridge(random_state=0),
    'ElasticNet': ElasticNet(random_state=0),
    'SGDRegressor' : SGDRegressor(random_state=0),
    'GradientBoostingRegress': GradientBoostingRegressor(random_state=0),
    'AdaBoostRegressor': AdaBoostRegressor(random_state = 0),
    'SVR': SVR(C=1.0, epsilon=0.2)
}
```

```python
for name_e, estimator in estimators.items():
    print(name_e)
    model = make_pipeline(RobustScaler(), estimator)
    evaluation(model, hypothesis=hypothesis[-1])
```

I'll chose GradientBoostingRegress

## Optimisation 

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
X_train, y_train = preprocessing(trainset, hypothesis=hypothesis[-1])
X_test, y_test = preprocessing(testset, hypothesis=hypothesis[-1])
```

```python
model = make_pipeline(RobustScaler(), GradientBoostingRegressor())
model
```

```python
hyper_params = {
    'gradientboostingregressor__learning_rate': [0.5, 0.1, 0.05, 0.01],
    'gradientboostingregressor__n_estimators': [32, 64, 100, 200],
    'gradientboostingregressor__max_depth': np.linspace(2, 16, 3, endpoint=True),
    'gradientboostingregressor__min_samples_split': np.linspace(0.1, 1.0, 3, endpoint=True),
    'gradientboostingregressor__max_features': np.linspace(0.1, 1.0, 3, endpoint=True)
}
```

```python
grid = GridSearchCV(model, hyper_params, scoring='neg_mean_squared_error', cv=4)

grid.fit(X_train, y_train)

print(grid.best_params_)
```

```python
print('Final score : ' + str(np.sqrt(-grid.best_score_)))
```
